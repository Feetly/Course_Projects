```python
def fn(n):
    ans,ctr=0,1
    for i in bin(n)[2:][::-1]:
        if i is '1': ans+=2*ctr
        ctr*=2
    return ans-ctr+1
fn(100)
```

```python
ans,ctr=0,1
def fn(n):
    global ans,ctr
    if n is 1 : print(ans+1)
    else :
        if n%2 : ans+=2*ctr
        n = n//2
        ctr*=2
        fn(n)      
fn(100)
```

```python
import numpy as np
no = int(input("No. of person in circle : "))
print("No. of last man alive : ",int(((np.array([2**i for i in range(1,int(np.log(no)/np.log(2)) + 1)]))*(np.array([int(i) for i in bin(no)[2:][::-1]][:-1]))).sum() + 1))  
```
